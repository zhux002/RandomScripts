```python
import random 
import matplotlib.pyplot as plt
import numpy as np
```

```python
t_final = 1000
delta_t = 0.01
n = int(t_final/delta_t)
alpha_scaling = 1
alpha = alpha_scaling*delta_t
E = 10
```

```python
y = np.zeros(n)
y[0] = 1

for i in range(1,n):
    y[i] = y[i-1] + alpha*E*y[i-1]*(1-y[i-1]**2) + np.sqrt(alpha/2)*random.gauss(0,1)
```

```python
t = np.linspace(0,t_final,n)
plt.plot(t,y)
```
