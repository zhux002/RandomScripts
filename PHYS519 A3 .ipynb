```python
import random 
import matplotlib.pyplot as plt
import numpy as np
```

```python
t_final = 1000
delta_t = 0.01
n = int(t_final/delta_t)
alpha_scaling = 1
alpha = alpha_scaling*delta_t
E = 1
```

```python
y = np.zeros(n)
y[0] = 1

for i in range(1,n):
    y[i] = y[i-1] + alpha*E*y[i-1]*(1-y[i-1]**2) + np.sqrt(alpha/2)*random.gauss(0,1)
```

```python
def normalize(y,dx):
    return y/np.sum(y*dx)
```

```python
y_boltzmann = np.linspace(-1.5,1.5,100)
p_boltzmann = np.exp(-E*(1-y_boltzmann**2)**2)
dx_theory = y_boltzmann[1]-y_boltzmann[0]
p_boltzmann_normalized = normalize(p_boltzmann,dx_theory)
plt.plot(y_boltzmann,p_boltzmann_normalized)
plt.hist(y,density=True,bins=40)
plt.xlabel('y')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of y for E = %.2f'%(E))
```

```python
t = np.linspace(0,t_final,n)
plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Trajectory of y for E = %.2f'%(E))
```

```python
t_final = 1000
delta_t = 0.01
n = int(t_final/delta_t)
alpha_scaling = 1
alpha = alpha_scaling*delta_t

transitions = []
E = np.linspace(1,30,10)

for j in E:
    
    y = np.zeros(n)
    transitions_temp = 0
    
    y[0] = 1

    for i in range(1,n):
        y[i] = y[i-1] + alpha*j*y[i-1]*(1-y[i-1]**2) + np.sqrt(alpha/2)*random.gauss(0,1)
    
    for k in range(1,n):
        if (y[k] > 0 and y[k-1] < 0) or (y[k] < 0 and y[k-1] > 0):
            transitions_temp += 1
    transitions.append(transitions_temp)
```

```python
transitions
```

```python
current = np.array(transitions)/t_final
current
```

```python
x = np.exp(-E)
plt.plot(x,current)
plt.xlabel('$e^{-E}$')
plt.ylabel('Current (transitions/time)')
plt.title('Current vs. $e^{-E}$')
```
